```python
# 한글 폰트 설치
!sudo apt-get install -y fonts-nanum
# 시스템 폰트 캐시 재구성
!sudo fc-cache -fv

# Matplotlib 런타임 캐시 삭제 (Colab 환경에 따라 파일 경로가 다를 수 있음)
# 캐시를 삭제하면 Matplotlib이 다시 시작할 때 폰트를 재탐색합니다.
!rm -rf ~/.cache/matplotlib

# Matplotlib 라이브러리를 재로드하고 폰트 캐시를 명시적으로 다시 빌드
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 나눔바른고딕 폰트 경로 찾기 (확실하게 폰트를 찾도록)
# 폰트 캐시가 삭제된 후에는 findfont가 폰트를 다시 찾을 수 있습니다.
font_path = fm.findfont(fm.FontProperties(family='NanumBarunGothic'))
print(f"NanumBarunGothic 폰트 경로: {font_path}")

# 폰트 설정 재적용
plt.rc('font', family='NanumBarunGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

print("Matplotlib 폰트 설정이 완료되었습니다. 이제 한글이 제대로 표시될 것입니다.")
```
